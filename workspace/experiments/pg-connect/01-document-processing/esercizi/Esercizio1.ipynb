# Acquisizione testo

Acquisisco il contenuto testuale (url e url1 sono due variabili che contengono l'indirizzo Web del txt da scaricare):

```python
from urllib import request
import ssl
context = ssl._create_unverified_context()
url = "https://www.gutenberg.org/files/2554/2554-0.txt"
url1 = "https://www.gutenberg.org/files/66419/66419-0.txt"
response = request.urlopen(url1,context=context)
text = response.read().decode('utf8')
print(text)
```

E' possibile caricare anche il testo di altre sorgenti dati utilizzando appositi package. Ad esempio estrarre il testo da un file PDF nel vostro file system.

# Preprocessing

```python
import nltk
```

```python
import nltk
nltk.download('book')

tokens = nltk.word_tokenize(text)
print( tokens)


tagged=nltk.pos_tag(tokens)

#print(tagged)
#
from nltk.corpus import stopwords

tokens2=set()

for t in tokens:
   if not t in stopwords.words('english'):
       tt = nltk.pos_tag([t])
       print(tt)
       if tt[0][1][0:2] == 'NN':
           tokens2.add(t)

print( "Tokens after elimination of stopwords and noun selection:")
print( tokens2)


```

```python
from nltk.stem.porter import PorterStemmer

porter = PorterStemmer()


for t in tokens2:
   print( porter.stem(t))
```
